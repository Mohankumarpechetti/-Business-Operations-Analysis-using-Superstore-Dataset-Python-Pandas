### Stage 4 Tasks: Data Cleaning & Transformation

```python
 # Handle missing values in the postal_code column.
df['Postal Code'].isna().sum()
```

```python
# Convert order_date and ship_date to datetime format.
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
# Create a new column profit_margin = profit / sales.
df['Profit Margin'] = df['Profit']/df['Sales']
```

```python
 # Create a new column shipping_delay = ship_date - order_date.
df['Shipping Delay'] = df['Ship Date'] - df['Order Date']
```

```python
# Bin the sales column into 3 categories: Low, Medium, and High
# Create new column for sales category
df['sales_category'] = df['Sales'].apply(lambda x: 'Low' if x < 500 else 'Medium' if x<1000 else 'High')
```

```python
df['sales_category'].value_counts()
```

### Stage 5 Tasks: Merging & Joining

```python
# Simulate Orders Table
orders = df[['Order ID', 'Customer ID', 'Product ID', 'Order Date', 'Sales', 'Profit']]

# Simulate Customers Table
customers = df[['Customer ID', 'Customer Name', 'Segment', 'Region']].drop_duplicates()

# Simulate Products Table
products = df[['Product ID', 'Product Name', 'Category', 'Sub-Category']].drop_duplicates()
```

```python
 # Merge the Orders table with the Customers table using customer_id.
```

```python
orders_customers = pd.merge(orders, customers, on='Customer ID', how='inner')
```

```python
orders_products = pd.merge(orders, products, on='Product ID', how='inner')
```

```python
left_join  = pd.merge(orders, customers, on='Customer ID', how='left')
right_join = pd.merge(orders, customers, on='Customer ID', how='right')
outer_join = pd.merge(orders, customers, on='Customer ID', how='outer')
```

```python
merged = pd.merge(orders, products, on='Product ID', how='inner')
merged.groupby('Category')['Sales'].sum().reset_index()
```

```python
merged_all = pd.merge(orders, products, on='Product ID', how='inner')
merged_all.groupby('Category')['Customer ID'].nunique().reset_index(name='unique_customers')
```

### Stage 7: Advanced Analysis & Insights

```python
# Set order_date as the index and sort the DataFrame by date.

# Make a copy to avoid affecting original df
df_time = df.copy()

# Set order_date as index
df_time.set_index('Order Date', inplace=True)

# Sort the index by date
df_time.sort_index(inplace=True)

# Preview the result
df_time.head()
```

```python
# Resample the data to show monthly total sales.

monthly_sales=df_time['Sales'].resample('ME').sum()
monthly_sales
```

```python
# Find the month with the highest profit
df_time['Profit'].resample('ME').sum().sort_values(ascending=False).head(1)
```

```python
 # Plot the monthly sales trend using Matplotlib or Pandas
import matplotlib.pyplot as plt

# Plot monthly sales trend
monthly_sales.plot(figsize=(12, 6), color='teal', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Create a new column called order_month in YYYY-MM format.
# Then group by order_month and category to analyze category-wise sales trends over time.

# Step 1: Create a new column for 'order_month'
df['order_month'] = df['Order Date'].dt.to_period('M').astype(str)

# Step 2: Group by order_month and category, then sum sales
monthly_category_sales = df.groupby(['order_month', 'Category'])['Sales'].sum().reset_index()

# Preview the result
print(monthly_category_sales.head())
```
